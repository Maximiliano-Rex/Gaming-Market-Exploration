```python
import pandas as pd
gaming_data = pd.read_csv("games-features-edit.csv")
gaming_data
```

```python
gaming_data.describe()
```

```python
gaming_data = gaming_data[gaming_data['PriceInitial'] <= 60.0]
gaming_data = gaming_data[gaming_data['Metacritic'] > 10.0]  
gaming_data = gaming_data[gaming_data['RecommendationCount'] > 10.0] 
```

```python
gaming_data.describe()
```

```python
import plotly.express as px
gaming_data = gaming_data[gaming_data['Metacritic'] > 90.0] 
fig = px.scatter(gaming_data, x="PriceInitial", y="Metacritic", trendline="ols", text="ResponseName")
fig.show()
```

```python
import plotly.express as px

fig = px.scatter(gaming_data, x="RecommendationCount", y="Metacritic", trendline="ols")
fig.show()
```

```python
import plotly.express as px
gaming_data = gaming_data[gaming_data['RecommendationCount'] > 10.0] 
fig = px.scatter(gaming_data, x="RecommendationCount", y="Metacritic", trendline="ols")
fig.show()
```

```python
gaming_data = gaming_data[gaming_data['RecommendationCount'] > 2000.0] 
gaming_data = gaming_data[gaming_data['PriceInitial'] <= 60.0]
gaming_data = gaming_data[gaming_data['Metacritic'] > 10.0]  
fig = px.scatter(gaming_data, x='RecommendationCount', y='Metacritic', 
                 trendline="ols", trendline_options=dict(log_x=True),
                 title="Log-transformed fit on linear axes",
                 trendline_color_override="black")
fig.show()
```
